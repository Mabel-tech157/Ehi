```python
# Determining the Genres using for loop
list_genres = mv.genres.apply(lambda x: x.split(","))
genres = []

for row in list_genres:
    row = [genre.strip() for genre in row]
    for genre in row:
        if genre not in genres:
            genres.append(genre)

for genre in genres:
    mv[genre] = mv['genres'].str.contains(genre).astype(int)

mv[genres]
```
